```python
from torch.utils.data import Dataset, DataLoader

class MyDataset(Dataset):
    def __init__(self,df):
        self.df=df
    def __len__(self):
        return len(self.df)
    def get_sample_dic(self, idx):
        diz=dict(self.df.iloc[idx])
        return diz
    
    def __getitem__(self,idx):
        return self.get_sample_dic(idx)
    
dataset=MyDataset(df)
sample=dataset[0]
print(sample)
print(type(sample))
print(sample.values())
```
