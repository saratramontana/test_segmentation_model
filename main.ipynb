```python
import segmentation_models_pytorch as smp


aux_params = {
    'pooling': 'avg',             # one of 'avg' or 'max'
    'dropout': 0.5,              # dropout ratio, default is None
    'activation': 'sigmoid',      # activation function, default is None
    'classes': 2,                # define number of output labels
}
model = smp.Unet(
    encoder_name="resnet34",        # choose encoder, e.g. mobilenet_v2 or efficientnet-b7
    encoder_weights="imagenet",     # use `imagenet` pre-trained weights for encoder initialization
    in_channels=3,                  # model input channels (1 for gray-scale images, 3 for RGB, etc.)
    classes=2,  
    aux_params=aux_params                 # model output channels (number of classes in your dataset)
)
model
```

```python
import torch

fake_batch_img =torch.rand(size=(16, 3, 256, 256))

fake_batch_img
```

```python
seg_logits, cls_logits = model(fake_batch_img)

```
