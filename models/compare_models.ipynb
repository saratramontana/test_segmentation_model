```python
baseline_df = pd.read_csv("results/baseline_results.csv")
acsnet_df = pd.read_csv("results/acsnet_results.csv")
ovamta_df = pd.read_csv("results/ovamta_results.csv")
```

```python
baseline_best = baseline_df.sort_values("val_loss").iloc[0]
acsnet_best = acsnet_df.sort_values("val_loss").iloc[0]

ovamta_diag_best = ovamta_df[
    ovamta_df["model"] == "ovamta_diag"
].sort_values("val_loss").iloc[0]
```

```python
comparison_df = pd.DataFrame([
    {
        "model": "baseline",
        "val_loss": baseline_best["val_loss"],
        "val_seg_dice": baseline_best["val_seg_dice"],
        "val_seg_iou": baseline_best["val_seg_iou"],
        "val_cls_f1": baseline_best["val_cls_f1"],
        "val_cls_auc": baseline_best["val_cls_auc"],
        "best_ckpt_path": baseline_best["best_ckpt_path"],
    },
    {
        "model": "acsnet",
        "val_loss": acsnet_best["val_loss"],
        "val_seg_dice": acsnet_best["val_seg_dice"],
        "val_seg_iou": acsnet_best["val_seg_iou"],
        "val_cls_f1": acsnet_best["val_cls_f1"],
        "val_cls_auc": acsnet_best["val_cls_auc"],
        "best_ckpt_path": acsnet_best["best_ckpt_path"],
    },
    {
        "model": "ovamta",
        "val_loss": ovamta_diag_best["val_loss"],
        "val_seg_dice": ovamta_diag_best["val_seg_dice"],
        "val_seg_iou": ovamta_diag_best["val_seg_iou"],
        "val_cls_f1": ovamta_diag_best["val_cls_f1"],
        "val_cls_auc": ovamta_diag_best["val_cls_auc"],
        "best_ckpt_path": ovamta_diag_best["best_ckpt_path"],
    },
])

comparison_df
```
